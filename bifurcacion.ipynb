```python
import matplotlib.pylab as plt
import numpy as np
from IPython.display import clear_output
```

```python
%%bash
c++ bifurcacion.cpp -o bi.out
./bi.out 100000 > bi.dat
```

```python
data = np.loadtxt("bi.dat")
thetas = data[:,0]
fuerzas = data[:,1]

plt.figure(figsize=(12,10)),
plt.scatter(fuerzas,thetas,s=4,c='y')
plt.xlabel('Fuerza (N)')
plt.ylabel('$\\Theta$')
plt.title('Bifurcacion para un Oscilador Armonico')
plt.grid()
plt.savefig("bifurcacion.png")
```
